## Possible solution #1 (easy)

```python
for number in range(1,101): # remember, end point is exclusive
    if number % 2 == 0:
        if number % 5== 0:
            print("fizzbuzz")
        else:
            print("fizz")
    elif number % 5 == 0:
        print("buzz")
    else:
        print(number)
```
