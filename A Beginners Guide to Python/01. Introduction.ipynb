### My Personal Course Recomendations: 

> [An Introduction to Interactive Programming in Python (Parts 1 & 2)](www.coursera.org/learn/interactive-python-1), Rice University (Coursera).

This was the very first course programming I ever took. At the time I found it really hard, but I can't recommend this course highly enough. Each week you will be learning how to build old-school games like Asteroids, Pong, Blackjack and so on.
  
> [Introduction to Computer Science and Programming Using Python](https://www.edx.org/course/introduction-computer-science-mitx-6-00-1x-8), Massachusetts Institute of Technology (Edx)

Again, this is also a hard course, but just like the one above with enough persistence, blood, sweat and tears you will get through it!

> [The Design of Computer Programs](https://www.udacity.com/course/design-of-computer-programs--cs212), Peter Norvig (Udacity)

This is an **advanced** course, and if I am to be honest the first time I took it I struggled in places. But there is some really good stuff here and so therefore I'd recommend you make a little reminder to check this course out in six months time. 
   
### For Programming Puzzles...

There are a few websites out there, but my overall favourite is [codewars.com](https://www.codewars.com). I'd strongly recommend you sign up and give a few puzzles a go. If you want to **actually learn** how to program you simply have to get coding and this website will give you plenty of opportunities to practice. Moreover, the the ranking system should help give you a sense of accomplishment and progress as you beat harder and harder puzzles.

## HOMEWORK ASSIGNMENT

Your task for this week is to think of a cool project. If you had the skills and time, what would you build?

* A game?
* An Automaton that files your taxes?
* Cure for Cancer?

Okay okay, finding a cure of cancer would be a huge project, so we might want to tone down the ambition just a bit! But Python is actually pretty good for "Bioinformatics". Thus, maybe we can't find a cure for cancer, but writing software that can scan a DNA sequence looking abnormalities is more than doable! 

Likewise, we are never going to be able to make a game like *Call of Duty* by ourselves, but building a game like say *Final Fantasy III* certainly is achievable. In later lectures I’ll talk about how to build Chess and Minesweeper in Python. 

In short, What's your programming dream?  Once you've figured that out try to come up with something a bit more realistic. Hold onto this idea, **This is your motivation**. Learning Python is not going to be easy, there are going to be times where you fall and stumble and if you don’t have a good reason to get *back up* you may just *give up*.
To succeed you are going to need to be motivated and that is the idea behind this first homework:

> Find your motivation guys. Its bloody important!
